```python

# if
score = 90
grade = ''
if score >= 90:  # [>90]
  grade = 'A'
elif score >= 80:  # score < 90 and score >=80 [80, 90)
  grade = 'B'
elif score >= 70:  # [70, 80)
  grade = 'C'
elif score >= 60:  # [60, 70)
  grade = 'D'
else: # score < 60
  grade = 'E'
print('Grade: ', grade)
```

```python
# class
class Person:
  def __init__(self, name, age): # constructor (ham tao doi tuong)
    self.name = name
    self.age = age

  def myfunc(self):
    print('Hello my name is ', self.name)
    pass

p1 = Person('Alan', 20)
print(p1.name, p1.age)
p1.myfunc()
```

```python
class Vehicle:
  name = ""
  kind = "car"
  color = ""
  value = 100.0

  def __init__(self, name="", kind="car", color="", value=100.0):
    self.name = name
    self.kind = kind
    self.color = color
    self.value = value
    pass

  def description(self):
    desc_str = f'{self.name} is a {self.color} {self.kind} worth ${self.value:.2f}.'
    return desc_str
    pass

car1 = Vehicle()
car1.name = 'Fer'
car1.color = 'red'
car1.kind = 'convertible'
car1.value = 60000.0

car2 = Vehicle("Jump", 'blue', 'van', 10000.0)

print(car1.description())
print(car2.description())
```
